```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
```

```python
base_url = "http://oldiesbutgoodies-records.com/?mode=cate&cbid=239269&csid=1&sort=p&page={}"
#detail_page ="http://oldiesbutgoodies-records.com/?pid={}"
# a_side_tune_name = None
# a_side_performer = None
# b_side_tune_name = None
# b_side_performer = None
# label = None
# price = None
# condition = None
# comments = None
# a_side_tune_text_l = []
d_list = []
for i in range(89,99):
    url1 =base_url.format(i)
    r1 = requests.get(url1,timeout=1)
    r1.raise_for_status()
    soup1 = BeautifulSoup(r1.content,"lxml")

    tune_blocks =soup1.select("tr[valign=top]")
    print(url1)

    

    for tune in tune_blocks:
        inc_pid =tune.select_one("a").get("href")
        pid = inc_pid.replace("?pid=","")
        detail_page_url = f"http://oldiesbutgoodies-records.com/?pid={pid}"
        sleep(2)
        r2 = requests.get(detail_page_url,timeout=1)
        r2.raise_for_status()
        soup2 = BeautifulSoup(r2.content,"lxml")
        tune_block = soup2.select_one("dl.spec")
        print(detail_page_url)
        
        
        
        if tune_block != None:
            
            a_side_tune_text =soup2.select_one("dl.spec > dd > a").text
            a_side_tune_text_l = a_side_tune_text.strip().split("/")
            a_side_performer = a_side_tune_text_l[0]
            a_side_tune_name = a_side_tune_text_l[1]

            b_side_tune_text =soup2.select_one("dl.spec > dd:nth-of-type(2) > a").text
            b_side_tune_text_l = b_side_tune_text.strip().split("/")
            b_side_performer = b_side_tune_text_l[0]
            b_side_tune_name = b_side_tune_text_l[1]

            label =soup2.select_one("dl.spec > dd:nth-of-type(3)").text.strip()
            condition = soup2.select_one("dl.spec > dd:nth-of-type(4)").text.strip()
            price = soup2.select_one("td.sales > font ").text
            price = int(price.strip().replace("円(税込)","").replace(",","")) if "円(税込)" in price else price
            comments = soup2.select_one("dl.spec > dd:last-of-type").text.strip()
            d_list.append({
                "Label":label,"A-Side_Tune":a_side_tune_name,"A-Side_Performer":a_side_performer,
                "B-Side_Tune":b_side_tune_name,"B-Side_Performer":b_side_performer,"Condition":condition,
                "Price":price,"Comments":comments
            })
            print(a_side_tune_name,a_side_performer,b_side_tune_name,b_side_performer,label,condition,price)
            
            print(d_list[-1])
            print()
        else:
            a_side_tune_name =soup2.select_one("div.category_title > b").text
            a_side_performer = soup2.select_one("div.category_title > font > b").text
            b_side_tune_text =soup2.select_one("table[width='400'][border='0'] > tr:nth-of-type(1) > td").text
            b_side_tune_text_l = b_side_tune_text.strip().split("/")
            b_side_performer = b_side_tune_text_l[0]
            b_side_tune_name = b_side_tune_text_l[1]

            label =soup2.select_one("table[width='400'][border='0'] > tr:nth-of-type(4) > td").text.strip()
            condition = soup2.select_one("table[width='400'][border='0'] > tr:nth-of-type(5) > td").text.strip()
            price = soup2.select_one("table.spec > tr > td > font").text
            comments = soup2.select_one("table[width='400'][border='0'] > tr:nth-of-type(6) > td").text.strip()
            d_list.append({
                "Label":label,"A-Side_Tune":a_side_tune_name,"A-Side_Performer":a_side_performer,
                "B-Side_Tune":b_side_tune_name,"B-Side_Performer":b_side_performer,"Condition":condition,
                "Price":price,"Comments":comments
            })
            print(a_side_tune_name,a_side_performer,b_side_tune_name,b_side_performer,label,condition,price)
            
            

          
            print(d_list[-1])
            print()
print(len(d_list))

df = pd.DataFrame(d_list)
df.to_csv("obg_ska_rs_45.csv",index=False,encoding = "utf-8")



        # a_side_tune_name_l= a_side_tune_name.split("/")
        # for p,t in a_side_tune_name_l:


        # a_side_performer =a_side_tune_name[:e_index]
        # a_side_tune_name = a_side_tune_name[e_index:]
        # b_side_tune_name = soup2.select_one("dl.spec > dd:nth-of-type(2) > a").text
        # b_side_tune_name = b_side_tune_name.strip()
        # b_side_tune_name =b_side_tune_name.replace(" ","")
        # e_index2 = a_side_tune_name.find("/")
        # b_side_performer =b_side_tune_name[:e_index2]
        # b_side_tune_name = b_side_tune_name[e_index2:]
        # label_name =soup2.select_one("dt:nth-of-type(3)").text
        # str_price = soup2.select_one("p.price_detail").text
        # int_price = str_price.replace("円(税込)","").replace(",","")
        #print(f"A-side:{a_side_tune_name},B-side:{b_side_tune_name}")
        #print(a_side_performer,a_side_tune_name,b_side_performer,b_side_tune_name)
        # print(int(int_price))
```
