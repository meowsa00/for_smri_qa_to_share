```python
from time import sleep,time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import re
import os
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--incognite')
#options.add_argument('--headless')
#user-agentの設定
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36")

driver_path = "C:/Users/2S9M7PAABJ/Documents/Programming/docker-python/tools/chromedriver.exe"

service = Service(executable_path=driver_path) # 2) executable_pathを指定
driver = webdriver.Chrome(service=service) #3) serviceを渡す
driver.implicitly_wait(10)
```

```python
url = "https://www.mizuhobank.co.jp/retail/takarakuji/check/numbers/backnumber/index.html"
driver.get(url)
sleep(3)
#1SeleniumでTopページからURL一覧を取得する

last_12months_links = driver.find_elements(By.CSS_SELECTOR,"tr.js-backnumber-temp-a>td:nth-of-type(1)>a[tabindex='200']")
back_number_links = driver.find_elements(By.CSS_SELECTOR,"tr.js-backnumber-temp-b>td>a[tabindex='200']")
urls = [atag.get_attribute("href") for atag in last_12months_links + back_number_links]

dir_path = os.path.dirname(os.path.abspath("__file__"))

for i,detail_url in enumerate(urls,start=1):
    print("="*30,i,"="*30)
    print(detail_url)
    driver.get(detail_url)
    sleep(3)
    HTML_source = driver.page_source
    title = re.sub(r'[\\/:*?"<>|]+','',driver.title)
    title = title.replace(" ","")
    title = title.replace("　","")
    filename = os.path.join(dir_path,"ensyu4_HTML_2",f"{title}.html")
    print(filename)
    with open(filename,"w",encoding = "UTF-8") as f:
        f.write(HTML_source)
sleep(3)

    
    
    
    
    
    


```
